```python
import pandas as pd 
```

```python
df=pd.read_csv('data/cars.csv')
df.head()
```

```python
df['Weight']>3500
```

```python
#Apply 
df.loc[df['Weight']>3500]
```

```python
#Apply conditions are get required
df.loc[df['Weight']>3500][['Car','Origin']]
```

```python
#Multiple Condition
df.loc[(df['Origin']=='Europe')&(df['Weight']>3500)]
```

## Between Function

```python
df.loc[df['Weight'].between(3000,3100)]
```

## Filter like Values

```python
df.loc[df['Car'].str.contains('Ford')]
```

Like using Index

```python
df_car=df.set_index('Car')
df_car.head()
```

```python
df_car .filter(like='Ford',axis=0)
```
