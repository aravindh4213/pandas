Functions

```python
import pandas as pd
```

```python
df=pd.read_csv('data/cars.csv')
df.head()
```

1) Apply Functions

```python
def convert_to_watts(hp:float):
    return hp*0.745
```

```python
df['watts']=df['Horsepower'].apply(convert_to_watts)
df.head()
```

2) Assign Functions

```python
df=df.assign(watts2=convert_to_watts(df['Horsepower']))
df.head()
```

Tranform

```python
df['watts3']=df['Horsepower'].transform(convert_to_watts)
df.head()
```

5)Lambda Function

```python
def convert_to_watts(hp:float):
    return hp*0.745
```

```python
lambda hp:hp*0.745
```

```python
df['lambda']=df['Horsepower'].apply(lambda hp:hp*0.745)
df.head()
```

```python
df['lambda']=df['Horsepower'].transform(lambda hp:hp*0.745)
df.head()
```

```python
df.assign(lassign=lambda df:df['Horsepower']*0.745)
df.head()
```

```python
df.assign(
    mwatt1=lambda df:df['Horsepower']*0.745,
    disp=lambda df:df['Displacement']/1000
)
df.head
```

6) Pipe function

```python
df['Pipe']=df['Horsepower'].pipe(lambda x:x*0.745).pipe(lambda x :round(x,1))
df.head()
```
